File Organization 


```python
import os

def organize_files(source_dir, destination_dir):
    """Organizes files based on their extensions into subdirectories.

    Args:
        source_dir (str): The directory containing the files to be organized.
        destination_dir (str): The directory where the organized files will be placed.
    """

    for filename in os.listdir(source_dir):
        filepath = os.path.join(source_dir, filename)
        extension = filename.split('.')[-1]
        subdirectory = os.path.join(destination_dir, extension)
        if not os.path.exists(subdirectory):
            os.makedirs(subdirectory)
        os.rename(filepath, os.path.join(subdirectory, filename))
        print(f"Moved {filename} to {subdirectory}")

# Example usage:
source_dir = "downloads"
destination_dir = "organized_files"
organize_files(source_dir, destination_dir)
```

```python
import os

def return_files_to_downloads(source_dir, destination_dir):
    """Moves files from organized subdirectories back to the Downloads directory.

    Args:
        source_dir (str): The directory containing the organized files.
        destination_dir (str): The original Downloads directory.
    """

    for subdirectory in os.listdir(source_dir):
        subdirectory_path = os.path.join(source_dir, subdirectory)
        if os.path.isdir(subdirectory_path):
            for filename in os.listdir(subdirectory_path):
                filepath = os.path.join(subdirectory_path, filename)
                target_path = os.path.join(destination_dir, filename)
                os.rename(filepath, target_path)
                print(f"Moved {filename} back to {destination_dir}")

# Example usage:
source_dir = "organized_files"
destination_dir = "downloads"
return_files_to_downloads(source_dir, destination_dir)
```
